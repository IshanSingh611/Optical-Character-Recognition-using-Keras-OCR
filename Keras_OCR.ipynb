```python
import keras_ocr
import matplotlib.pyplot as plt
```

```python
# keras-ocr will automatically download pretrained
# weights for the detector and recognizer.
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
image=keras_ocr.tools.read('https://www.opentext.com/file_source/OpenText/en_US/PNG/OT_ShareImage_twitter.png')
```

```python
type(image)
```

```python

plt.imshow(image)
```

```python
# Each predictions is a list of
# (word, box) tuples.
predictions = pipeline.recognize([image])
```

```python
keras_ocr.tools.drawAnnotations(image,predictions=predictions[0])
```
